```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
#Save the new files in a folder
save_name = "Clean_Mean_Files"
os.makedirs(save_name, exist_ok = True)

save_path = save_name + "/"
```

```python
path = "Clean_Files/"

file_1 = "clean_gcbu_dnc_stim_stage.xlsx"
#Nothing to be done
file_2 = "clean_monthly_production.xlsx"
file_3 = "clean_gcbu_dnc_recompletions.xlsx"
file_4 = "clean_mcbu_dnc_timelog_detail_reformatted.xlsx"
file_5 = "TO_TACC_STM_STG.xlsx"
file_6 = "TO_TACC_RCOMPL.xlsx"
```

```python
df_1 = pd.read_excel(path + file_1)
df_2 = pd.read_excel(path + file_2)
df_3 = pd.read_excel(path + file_3)
df_4 = pd.read_excel(path + file_4)
df_5 = pd.read_excel("UT_BestFrac_Dataset1/" + file_5)
df_6 = pd.read_excel("UT_BestFrac_Dataset1/" + file_6)
```

```python
df_1
```

```python
#Replace NaNs with mean of the column for file 1
quant_df = df_1.loc[:, df_1.dtypes == float]
impute = lambda col: col.fillna(col.mean())
imputed_df_1 = quant_df.apply(impute, axis=0)
```

```python
imputed_df_1.columns
```

```python
#Merge means back to original file
dropped_1 = df_1.drop(columns = imputed_df_1.columns)
result = pd.concat([dropped_1, imputed_df_1], axis = 1, sort = False)
```

```python
result
```

```python
imputed_df_1
```

```python
result.to_excel(save_path + "clean_mean" + file_1,index=False)
```
