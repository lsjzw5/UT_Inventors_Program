```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
#Save the new files in a folder
save_name = "Clean_Mean_Files"
os.makedirs(save_name, exist_ok = True)

save_path = save_name + "/"
```

```python
path = "Clean_Files/"

file_1 = "clean_gcbu_dnc_stim_stage.xlsx"
#Nothing to be done
file_2 = "clean_monthly_production.xlsx"
file_3 = "clean_gcbu_dnc_recompletions.xlsx"
```

```python
df_1 = pd.read_excel(path + file_1)
df_2 = pd.read_excel(path + file_2)
df_3 = pd.read_excel(path + file_3)
```

```python
df_1
```

```python
#Replace NaNs with mean of the column for file 1
quant_df = df_1.loc[:, df_1.dtypes == float]
impute = lambda col: col.fillna(col.mean())
imputed_df_1 = quant_df.apply(impute, axis=0)
```

```python
imputed_df_1.columns
```

```python
imputed_df_1
```

```python
imputed_df_1.to_excel(save_path + "clean_mean" + file_1,index=False)
```

```python
#Replace NaNs with mean of the column for file 3
quant_df = df_3.loc[:, df_3.dtypes == float]
impute = lambda col: col.fillna(col.mean())
imputed_df_3 = quant_df.apply(impute, axis=0)
```

```python
imputed_df_3.columns
```

```python
imputed_df_3
```

```python
imputed_df_3.to_excel(save_path + "clean_mean" + file_3,index=False)
```
